# Definition: 

Linear regression is a statistical analysis method that determines a linear relationship by finding the line of best fit between a regressor (independent, x axis) and the regressand (dependent, y axis). Two main forms of linear regression are SLR (Simple linear Regression) and MLR (Multiple linear Regression).
